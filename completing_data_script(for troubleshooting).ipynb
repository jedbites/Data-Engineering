```python
#import needed package
import asyncio
import httpx
import pandas as pd
import secret
import refresh
from time import perf_counter
```

```python
df = pd.read_excel('scraped_ids.xlsx')
```

```python
df
```

```python
starting_num = 62869
list_num = []
```

```python
df_list = list(df['id'])
```

```python
while starting_num != 103729:
    if starting_num not in df_list:
        list_num.append(starting_num)
        starting_num += 1
    else:
        starting_num += 1
```

```python
list_num
```

```python
#authtokens and headers
id_token = refresh.refreshed()
headers = {"content-type":"application/json", "x-api-key": secret.api_key, "id-token": id_token}
```

```python
#api calls
sem = asyncio.Semaphore(100)
#scrape all id

async def get_result(vinny_id:int):
    async with sem:
        async with httpx.AsyncClient() as client:
            r = await client.get(f'https://bps-world.vincere.io/api/v2/candidate/{vinny_id}/industries', headers=headers)
            if r.status_code == 200:
                data = [{**d, 'vinny_id': vinny_id} for d in r.json()]
                results.append(data)
                return
            else:
                result_error[vinny_id] = r.status_code
                return

#trigger main function
async def main():
    tasks = []
    #gen_id = df['vinny_id']
    for i in list_num:
        tasks.append(get_result(i))
    await asyncio.gather(*tasks, return_exceptions=True)

#future container for finding errors and stuff
results = []

#result errors
result_error = {}
```

```python
#trigger the whole function
start = perf_counter()
await main()
stop = perf_counter()
print('time taken:', stop - start, ' seconds')
```
