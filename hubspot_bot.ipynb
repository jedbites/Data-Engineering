```python
#modules needed for selenium
import requests
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
from random import randint
```

```python
def login(email: str, password: str) -> None:
    #Login to your HS account
    user_email = driver.find_element(By.XPATH, '//*[@id="username"]').send_keys(email)
    user_password = driver.find_element(By.XPATH, '//*[@id="password"]').send_keys(password)
    sign_in_button = driver.find_element(By.XPATH, '//*[@id="loginBtn"]').click()
    return

def contacts_page() -> None:
    #Navigate to the contacts home page
    #driver.implicitly_wait(3)
    navigate_contacts_hover = driver.find_element(By.XPATH, './/*[@id="nav-primary-contacts-branch"]').click(driver.implicitly_wait(3))
    click_contacts_on_menu_dropdown = driver.find_element(By.XPATH, '//*[@id="nav-secondary-contacts"]').click()
    return

def get_url(hs_id: int) -> str:
    #url of contacts
    url = "https://app.hubspot.com/contacts/391079/contact/{}".format(hs_id)
    return url
```

```python
def main():
    df = pd.read_excel('uk_contacts.xlsx')
    df_record_id = df['Record ID']
    
    #navigate to contacts page
    #driver.implicitly_wait(3)
    #contacts_page()
    
    #waits for the elements to load
    driver.implicitly_wait(1)
    
    #iterate through record ids and validates it
    for i in df_record_id:
        url = get_url(i)
        driver.get(url)
        driver.implicitly_wait(2)
        
        #select phone number data field
        try:
            edit_phone_number = driver.find_element(By.XPATH, '//*[@data-selenium-test="property-input-phone"]').click()
            driver.implicitly_wait(1)
        except:
            pass
        
        #if already validated -> pass
        try:
            if driver.find_element(By.XPATH, '//i18n-string[contains(text(), "Validated")]'):
                pass
        except NoSuchElementException:
            pass
        
        #if the phone number is invalid or in wrong format
        try:
            if driver.find_element(By.XPATH, '//*[@data-unit-test="error-number-invalid"]'):
                driver.implicitly_wait(1)
                driver.find_element(By.XPATH, '//*[@class="phone-number-input"]/div').click()
                driver.implicitly_wait(1)
                driver.find_element(By.XPATH, '//*[@placeholder="Search"]').send_keys("United Kingdom", Keys.ENTER)
                driver.implicitly_wait(1)
                driver.find_element(By.XPATH, '//*[@class="uiButton private-button private-button--tertiary private-button--xs private-editable-control__save private-button--non-link"]').click()

                #to wait for atleast 3 secs
                time.sleep(1)
                #Save changes
                if driver.find_element(By.XPATH, '//button[@data-selenium-test="UniversalSaveBar-button-primary"]'):
                    driver.find_element(By.XPATH, '//button[@data-selenium-test="UniversalSaveBar-button-primary"]').click()
        except NoSuchElementException:
            pass
        
        #if not validated
        try:
            if driver.find_element(By.XPATH, '//i18n-string[contains(text(), "Apply number formatting")]'):
                driver.find_element(By.XPATH, '//i18n-string[contains(text(), "Apply number formatting")]').click()
                #select country code field
                driver.implicitly_wait(1)
                driver.find_element(By.XPATH, '//*[@class="phone-number-input"]/div').click()
                #enter United States
                driver.implicitly_wait(1)
                driver.find_element(By.XPATH, '//*[@placeholder="Search"]').send_keys("United Kingdom", Keys.ENTER)
                #Save validation data
                driver.implicitly_wait(1)
                driver.find_element(By.XPATH, '//*[@class="uiButton private-button private-button--tertiary private-button--xs private-editable-control__save private-button--non-link"]').click()
                
                #to wait for atleast 3 secs
                time.sleep(1)
                #Save changes
                if driver.find_element(By.XPATH, '//button[@data-selenium-test="UniversalSaveBar-button-primary"]'):
                    driver.find_element(By.XPATH, '//button[@data-selenium-test="UniversalSaveBar-button-primary"]').click()
        except NoSuchElementException:
            pass
        
        #if the phone number needs to select country code but doesn't gave a value
        try:
            if driver.find_element(By.XPATH, '//i18n-string[contains(text(), "Select a country code and enter a phone number.")]'):
                pass
        except NoSuchElementException:
            pass
```
