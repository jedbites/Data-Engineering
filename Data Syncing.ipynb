```python
count = 0
```

```python
for id in vinny_contacts['vinny_id']:
    data = {'data':[]}
    for column in columns:
        data['data'].append({'field_key':str(column),'field_value':str(vinny_contacts[column][count])})
    r = requests.put(get_url(id,'candidate','customfields'), headers=headers, json=data)
    if r.status_code == 401:
        print(r.json())
        print(id)
        break
    elif r.status_code == 429:
        print(r.json())
        print(id)
        break
    elif r.status_code == 400:
        print(r.json())
        print(id)
        break
    vinny_contacts.drop(index=vinny_contacts.index[0], axis=0, inplace=True)
    print(r.status_code)
    count += 1
```
